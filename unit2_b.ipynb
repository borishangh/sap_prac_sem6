```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
def compute_likelihood(N, M, f_values):
    likelihood = comb(N, M) * (f_values ** M) * ((1 - f_values) ** (N - M))
    return likelihood

N, M = 10, 7

f_values = np.linspace(0, 1, 1000)
likelihood = compute_likelihood(N, M, f_values)

mle_f = f_values[np.argmax(likelihood)]

print(f"Maximum Likelihood Estimate (MLE) for f: {mle_f:.3f}")
plt.figure(figsize=(10,5))
plt.plot(f_values, likelihood, 'b-', label='Likelihood')
plt.axvline(mle_f, color='r', linestyle='--', label=f'MLE: f = {mle_f:.3f}')
plt.xlabel("Probability of heads (f)")
plt.ylabel("Likelihood")
plt.title(f"Likelihood Function for Binomial Data (N={N}, M={M})")
plt.legend()
plt.show()
```
