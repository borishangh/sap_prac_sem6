```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

def plot_beta_prior_posterior(N, M, a, b):
    
    # posterior parameters
    a_post = a + M
    b_post = b + (N - M)
    
    f = np.linspace(0, 1, 1000)
    
    prior = beta.pdf(f, a, b)
    posterior = beta.pdf(f, a_post, b_post)
    
    plt.plot(f, prior, 'b-', lw=2, label=f'Prior: Beta({a}, {b})')
    plt.plot(f, posterior, 'r-', lw=2, label=f'Posterior: Beta({a_post}, {b_post})')
    
    f_obs = M/N
    plt.axvline(f_obs, color='k', linestyle='--', 
                label=f'Observed: f = {f_obs:.3f}')
    
    plt.title('Prior and Posterior Distributions of Coin Bias')
    plt.xlabel('Probability of heads (f)')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

N, M = 10,7
a, b = 3, 2

plot_beta_prior_posterior(N, M, a, b)
```
